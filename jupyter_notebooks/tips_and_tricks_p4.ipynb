# Tips and Tricks with Jupyter Notebooks


## Part 4: Creating Interactive dashboards

Building interactive dashboards in Jupyter Notebooks can potentially be a long and complicated process.

Interactive dashboards in Jupyter won't work if you export them as static html files

***Top Tip: Understand your requirements***

Should you need to visualize your data with minimal data manipulation, it may be quicker to use data visualization tools like [PowerBI](https://powerbi.microsoft.com/en-us/) or [Tableau](https://www.tableau.com/). 

- Ultimately, it would use less coding!

***Consider some Cloud solutions :)***

Working with Jupyter, we may want to share our dashboards, but interactive forms of notebooks are not easily shareable.

[dataschool.io](https://www.dataschool.io/cloud-services-for-jupyter-notebook/) evaluates some cloud notebook tools with a breakdown of their considerations:

- Supported Languages
- Ability to install packages
- Interface similarity
- Sharing publicly
- Ability to collaborate

Example technologies: 

- [Binder](https://mybinder.org/)
- [Kaggle Kernels](https://www.kaggle.com/)
- [Google Colab](https://colab.research.google.com/notebooks/intro.ipynb#recent=true)
- [Microsoft Azure notebooks](https://notebooks.azure.com/#)
- [Databricks](https://databricks.com/)
- [EC2 hosted Jupyter Notebook instance on virtual machine](https://dzone.com/articles/running-your-jupyter-notebooks-on-the-cloud)


### Typical libraries for Jupyter Dashboards

- [Pandas](https://pandas.pydata.org/), Numpy
- [SciKit-Learn](https://scikit-learn.org/stable/index.html) - Machine Learning
- [Matplotlib](https://matplotlib.org/), [Seaborn](http://seaborn.pydata.org/), [Plotly](https://plotly.com/) - data visualization
- [ipywidgets](https://ipywidgets.readthedocs.io/en/stable/examples/Widget%20List.html) e.g. for dropdowns, sliders etc.
