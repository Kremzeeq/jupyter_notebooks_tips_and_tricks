# Tips and Tricks with Jupyter Notebooks


## Part 4: Creating Interactive dashboards

Building interactive dashboards in Jupyter Notebooks can potentially be a long and complicated process.

***Top Tip: Understand your requirements***

It may be ok for small projects just for you. However, it may be quicker to use a data visualization tool like [PowerBI](https://powerbi.microsoft.com/en-us/) or [Tableau](https://www.tableau.com/). 

- It will use less coding
- They can provide a more shareable solution
- You don't have to worry about installing requirements

If you are looking to collaborate with notebooks in a shared environment, consider a cloud solution like [Databricks](https://databricks.com/).

- Remember, interactive dashboards in Jupyter won't work if you export them as static html files

### Typical libraries for Jupyter Dashboards

- Pandas, Numpy
- SciKit-Learn - models
- Matplotlib, Seaborn, Plotly - data visualization
- [ipywidgets](https://ipywidgets.readthedocs.io/en/stable/examples/Widget%20List.html) e.g. for dropdowns, sliders etc.

### Dashboard example: Google Analytics Linear Regression Project

My [linear regression project](https://github.com/Kremzeeq/ga_linear_regression_project) presents some web traffic data in an interactive dashboard.

This is split into two sections:

**Section A**

- Shows Graphical representation for Linear Model based on training data using Seaborn.
- Predictor box provided so the number of Users can be entered to predict Sessions and Page Views.

**Section B**

- Provides an overview on web metrics, built using plotly which offers interactive elements for graphs beyond dropdown ipywidgets

This can been constructed with classes for controllers and views

```python
from PIL import Image               # to load images
from IPython.display import display # to display images

basewidth = 600
pil_img = Image.open('../assets/dashboard_section_a.png')
wpercent = (basewidth/float(pil_img.size[0]))
hsize = int((float(pil_img.size[1])*float(wpercent)))
pil_img = pil_img.resize((basewidth,hsize), Image.ANTIALIAS)

display(pil_img)
```
